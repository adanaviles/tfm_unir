```python
from scripts.config import SRC_DIR
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SRC_DIR
```

```python
train = pd.read_csv(SRC_DIR / "data/internal/train.csv")
logs = pd.read_csv(SRC_DIR / "data/internal/user_logs.csv", nrows=30000000)
logs = pd.merge(left=train,
                            right=logs,
                            how='left',
                            on='msno')
logs.head()
```

```python
logs.describe()
```

```python
def convert_date(yyyymmdd):
    if pd.isna(yyyymmdd):
        return np.nan
    yyyymmdd_str = str(int(yyyymmdd))  # Convert to integer first to remove decimal part, then back to string
    yyyy = yyyymmdd_str[:4]
    mm = yyyymmdd_str[4:6]
    dd = yyyymmdd_str[6:]
    return f"{yyyy}-{mm}-{dd}"
logs['date'] = logs['date'].apply(convert_date)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(311)
sns.distplot(logs['num_25'],hist=False)
plt.xlabel('num_25', fontsize=12)
plt.title("Distribución de num_25", fontsize=12)
plt.show()
```
